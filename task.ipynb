```python
#Подсчет торговых дней
tickers = []
tick = open('Tickers.txt')
tickers = [i.replace('\n', '') for i in tick.readlines()]
tick.close()
f=[]

tabl = pd.DataFrame(index = tickers)
goda = [i for i in range(2011, 2021)]

for y in goda:
    t = []
    for i in range(len(tickers)):
        tick = pd.read_csv(tickers[i] + '.csv')
        dat=[int(datetime.datetime.strptime(i, '%Y-%m-%d').strftime('%Y')) for i in tick['Date']]
        DATE=pd.Series(dat)
        data = (DATE >=y) & (DATE < (y + 1))
        t.append(len(tick[data]))
    tabl[str(y)] = t
```

```python
#Подсчет максимальных и минимальных скачков
tickers = []
tick = open('Tickers.txt')
tickers = [i.replace('\n', '') for i in tick.readlines()]
tick.close()
tabl = pd.DataFrame(index = tickers)
tabl_1 = pd.DataFrame(index = tickers)
goda = [i for i in range(2011, 2021)]
for y in goda:
    high = []
    low = []
    for i in range(len(tickers)):
        tick = pd.read_csv(tickers[i] + '.csv')
        dat=[int(datetime.datetime.strptime(i, '%Y-%m-%d').strftime('%Y')) for i in tick['Date']]
        DATE=pd.Series(dat)
        data = (DATE >=y) & (DATE < (y + 1))
        far=tick[data]
        s=(far['Close']-far['Open'])/far['Open']
        high.append(round(max(s)*100,3))
        low.append(round(min(s)*100,3))
    tabl[str(y)] = high
    tabl_1[str(y)] = low
a=[max(tabl.loc[i]) for i in tickers]
b=[min(tabl_1.loc[i]) for i in tickers]
tabl['Max']=a
tabl_1['Min']=b
```

```python
#Подсчет p-value для реальных данных за год
a, p_value, znach=[],[],[]
alpha1, alpha2 = 0.05, 0.01
count1, count2 = 0,0
tabl1 = pd.DataFrame(index = tickers)
for y in goda:
    t,a =[],[]
    for i in range(len(tickers)):
        tick = pd.read_csv(tickers[i] + '.csv')
        dat=[int(datetime.datetime.strptime(i, '%Y-%m-%d').strftime('%Y')) for i in tick['Date']]
        DATE=pd.Series(dat)
        data = (DATE >=y) & (DATE < (y + 1))
        f=tick[data]
        log=(list((f['Close']-f['Open'])/f['Open']))
        real_d=st.shapiro(log)
        a.append(round(real_d[1],3))
        p_value.append(round(real_d[1],3))
        if real_d[1]>alpha1:
            count1+=1
        if real_d[1]>alpha2:
            count2+=1
    tabl1[str(y)] = a
dolya005=count1/(len(goda)*len(tickers))
dolya001=count2/(len(goda)*len(tickers))
tabl1
```
